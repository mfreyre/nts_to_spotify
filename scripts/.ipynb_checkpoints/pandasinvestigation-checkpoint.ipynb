```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file_path = "spotify_RGA.csv"
# Read the CSV file into a pandas DataFrame
df = pd.read_csv(csv_file_path, header=0)

# View the first 5 rows of the DataFrame
print(df.head())
```

```python
valence_info = df[["Energy", "Valence"]]
```

```python
valence_info.head()
```

```python
# Get a list of column names excluding 'valence'
x_columns = ['Energy','Instrumentalness', 'Popularity', 'Danceability' ]

# Create a figure with subplots for each x column
fig, axes = plt.subplots(nrows=1, ncols=len(x_columns), figsize=(15, 5))

# Create scatter plots of valence vs each x column
for i, col in enumerate(x_columns):
    df.plot.scatter(x=col, y='Valence', ax=axes[i])
    axes[i].set_title(col)

# Adjust the spacing between subplots and display the figure
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot of valence vs danceability, labeling each dot with the song title
fig, ax = plt.subplots(figsize=(10, 7))
sample = df.sample(frac=0.25)
sample.plot.scatter(x='Danceability', y='Valence', ax=ax)

for i, row in sample.iterrows():
    ax.annotate(row['Track Name'], xy=(row['Danceability'], row['Valence']), ha='left', va='center', fontsize=8)

plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(df['Valence'], bins=20)

# Add axis labels and a title
ax.set_xlabel('Valence')
ax.set_ylabel('Count')
ax.set_title('Distribution of Valence')

plt.show()
```
